```python
a=int(input("Veuillez entrer un bloc N de 8 caratères binaires :"))
b=int(input("Veuillez entrer une permutation PI: "))
cle1=int(input("Veuillez entrer la première clé "))
cle2=int(input("Veuillez entrer la deuxième clé "))
tab=[]
mab=[]
perm=[]
cle11=[]
cle22=[]
m=0
n=0
r=0
y=0
w=0







for i in range(0,8,1):
    c=int(b/10**(7-i))
    d=c
    c=c-m
    tab.append(c)
    m=d*10
for j in range(0,8,1):
    f=int(a/10**(7-j))
    g=f
    f=f-n
    mab.append(f)
    n=g*10








print(f"Votre bloc N en liste donne :{mab}")
print(f"Votre permutation en liste donne : {tab}")
for i in range(0,8,1):
    k=tab[i]
    s=mab[k]
    perm.append(s)
print(f"La permutation PI appliquée sur le bloc N est PI(N) = {perm}")
Go=perm[:4]
Do=perm[4:]
print(F"Go={Go}")
print(F"Do={Do}")
print('-----------------------------------------')
print('Premier round')
pe=2013
petab=[2,0,1,3]
pe_Go=[]
D1=[]
for i in range(0,4,1):
    l=petab[i]
    t=Go[l]
    pe_Go.append(t)
print(f"La permutation 2013 sur Go nous donne P(Go)= {pe_Go}")
    
    
for j in range(0,4,1):
    x=int(cle2/10**(3-j))
    e=x
    x=x-y
    cle22.append(x)
    y=e*10
    
for i in range(0,4,1):
    q=int(cle1/10**(3-i))
    v=q
    q=q - w
    cle11.append(q)
    w=v*10

print(f"La première clé que vous avez soumise est {cle11}")
print(f"La deuxième clé que vous avez soumise est {cle22}")

for i in range(0,4,1):
    if pe_Go[i] != cle11[i]:
        t=1
    else:
        t=0
    D1.append(t)
print("D1 = P(Go) xor k1")

chris=[]
for i in range(0,4,1):
    if Go[i]==1 or cle11[i]==1:
        tr=1
    else:
        tr=0
    chris.append(tr)
print(f"Go v K1 ={chris}")

G1=[]
for i in range(0,4,1):
    if Do[i]!=chris[i]:
        th=1
    else:
        th=0
    G1.append(th)
print(f"G1 = Do xor (Go v k1)")   
print(f"D1 = {D1}")
print(f"G1 = {G1}")    

print('---------------------------------------------')
print("Deuxième round")


pe_G1=[]
for i in range(0,4,1):
    l=petab[i]
    t=G1[l]
    pe_G1.append(t)
print(f"P(G1)={pe_G1}")

D2=[]
for i in range(0,4,1):
    if pe_G1[i] != cle22[i]:
        th=1
    else:
        th=0
    D2.append(th)

env=[]
for i in range(0,4,1):
    if G1[i]==1 or cle22[i]==1:
        tr=1
    else:
        tr=0
    env.append(tr)
print(f"D2=P(G1) xor k2")    
print(f"G1 v K2 ={env}")
print(f"G2=D1 xor (G1 v K2)") 
G2=[]
for i in range(0,4,1):
    if D1[i] != env[i]:
        th=1
    else:
        th=0
    G2.append(th)
    
print(f"D2 = {D2}")    
print(f"G2= {G2}")

C=[]
C=G2+D2
print(f"La concaténation de G2 et D2 nous donne{C}")

from sympy.combinatorics import Permutation
p = Permutation(tab)
inv = p.__invert__()
print(f"L'inverse de la permutation PI est  {inv.array_form}") # [1, 3, 2, 0]

zuzu=[]
for i in range(0,8,1):
    k=inv.array_form[i]
    s=C[k]
    zuzu.append(s)
print("---------------------------------------------------------")
print(f"Le message chiffré est :{zuzu}")
```

# Algorithme de la génération des clés

```python
a=int(input("Veuillez entrer un bloc N de 8 caratères binaires :"))
b=int(input("Veuillez entrer une permutation H: "))
off=int(input("Veuillez entrer un décalage pour la première clé : "))
but=int(input("Veuillez entrer un décalage pour la deuxième clé : "))

#Veuillez entrer un décalge négatif pour le décalage de droite 

tab=[]
mab=[]
perm=[]

m=0
n=0
r=0
y=0
w=0
for i in range(0,8,1):
    c=int(b/10**(7-i))
    d=c
    c=c-m
    tab.append(c)
    m=d*10
for j in range(0,8,1):
    f=int(a/10**(7-j))
    g=f
    f=f-n
    mab.append(f)
    n=g*10

    
print(f"Votre bloc N en liste donne :{mab}")
print(f"Votre permutation en liste donne : {tab}")


for i in range(0,8,1):
    k=tab[i]
    s=mab[k]
    perm.append(s)
print(f"La permutation H appliquée sur N donne H(N) ={perm}")
k_1=perm[:4]
k_2=perm[4:]

print("En divisant le bloc H(N) en deux blocs de 4 bits, nous obtenons: ")
print(F"Le premier bloc K'1={k_1}")
print(F"Le second bloc K'2={k_2}")




k1=[]

for i in range(0,4,1):
    if k_1[i] != k_2[i]:
        t=1
    else:
        t=0
    k1.append(t)



k2=[]


for i in range(0,4,1):
    if k_1[i]==1 and k_2[i]==1:
        tr=1
    else:
        tr=0
    k2.append(tr)
    
print(f"k1 = K'1 xor K'2")
print(f"k2 = K'2 et K'1")
    
print(f"k1 = {k1}")
print(f"k2 ={k2}")


def decaler(list, offset):
    i = offset % len(list)
    return list[i:] + list[:i]
print("En appliquant les décalages entrés, nous avons : ")

x= decaler(k1, off)
y= decaler(k2, but)



print(f"key1= {x}")
print(f"key2 ={y}")
print(f"A la sortie, nous avons la paire de deux clés ({x}, {y})")
```
